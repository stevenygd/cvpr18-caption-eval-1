# Section 4.2

```python
def extract_results(filename, n_epoch):
    n_epoch -= 1
    f = open(filename, 'r')
    results_raw = f.read().split('\n')
    results_raw = results_raw[1:11] if len(results_raw) == 12 else results_raw[:10]
    results = np.zeros((10, 4), dtype=np.float32)
    for i in range(10):
        r = results_raw[i].split()
        for j in range(4):
            results[i, j] = r[j*2]
    g = np.mean(results[n_epoch, 0:2])
    h = results[n_epoch, -1]
    f.close()
    return h, g
```

```python
n_epoch = 10
exp_name = 'capability_da/'
save_path = './model/' + exp_name
n_layers = 1
hidden_size = 128
image_features = ['no_img', 'only_img', 'img']
models = ['concat', 'bilinear', 'mlp_1']

scores = np.zeros((3,3,2), dtype=np.float32)
for i in range(len(image_features)):
    for j in range(len(models)):
        filename = save_path + models[j] + '_' + image_features[i] + '_1_512_0.txt'
        print(filename)
        h, g = extract_results(filename, n_epoch)
        scores[i,j,0] = h
        scores[i,j,1] = g
print(scores)
```

# Sec. 4.3

```python
scores_alpha_old = np.load('./data/scores_alpha.npy', encoding='latin1').item()
scores_beta_old = np.load('./data/scores_beta.npy', encoding='latin1').item()
scores_gamma_old = np.load('./data/scores_gamma.npy', encoding='latin1').item()
```

```python
print(scores_alpha_old)
```

```python
def add_our_scores(scores, bilinear, linear, mlp, bilinear_da, linear_da, mlp_da):
    c = 0
    for i in range(0,101,5):
        scores[i]['bilinear'] = np.float(bilinear[c])
        scores[i]['linear'] = np.float(linear[c])
        scores[i]['mlp'] = np.float(mlp[c])
        scores[i]['bilinear_da'] = np.float(bilinear_da[c])
        scores[i]['linear_da'] = np.float(linear_da[c])
        scores[i]['mlp_da'] = np.float(mlp_da[c])
        c += 1
    return scores
```

```python
for test_type in ['alpha', 'beta', 'gamma']:
    bilinear = open('./model/robustness/bilinear_img_1_512_0_noda_%s_robustness.txt' % test_type).read().split()
    linear = open('./model/robustness/concat_img_1_512_0_noda_%s_robustness.txt' % test_type).read().split()
    mlp = open('./model/robustness/mlp_1_img_1_512_0_noda_%s_robustness.txt' % test_type).read().split()
    bilinear_da = open('./model/robustness/bilinear_img_1_512_0_%s_robustness.txt' % test_type).read().split()
    linear_da = open('./model/robustness/concat_img_1_512_0_%s_robustness.txt' % test_type).read().split()
    mlp_da = open('./model/robustness/mlp_1_img_1_512_0_%s_robustness.txt' % test_type).read().split()
    if test_type == 'alpha':
        scores_alpha = add_our_scores(scores_alpha_old, bilinear, linear, mlp, bilinear_da, linear_da, mlp_da)
    elif test_type == 'beta':
        scores_beta = add_our_scores(scores_beta_old, bilinear, linear, mlp, bilinear_da, linear_da, mlp_da)
    elif test_type == 'gamma':
        scores_gamma = add_our_scores(scores_gamma_old, bilinear, linear, mlp, bilinear_da, linear_da, mlp_da)
```

```python
def convert_to_np(scores, names, keys):
    scores_np = np.zeros((len(names), 21), np.float32)
    for i in range(len(names)):
        for j in range(len(keys)):
            scores_np[i, j] = scores[keys[j]][names[i]]
    return scores_np
```

```python
def normalization(scores):
    for i in range(scores.shape[0]):
        s0 = scores[i, 0]
        for j in range(scores.shape[1]):
            scores[i, j] /= s0
    return scores
```

```python
names = ['Bleu_1', 'Bleu_2', 'Bleu_3', 'Bleu_4', 'METEOR', 'ROUGE_L', 'CIDEr', 'Spice',
         'bilinear', 'linear', 'mlp', 'bilinear_da', 'linear_da', 'mlp_da']
keys = range(0,101,5)

scores_alpha_all = normalization(convert_to_np(scores_alpha, names, keys))
scores_beta_all = normalization(convert_to_np(scores_beta, names, keys))
scores_gamma_all = normalization(convert_to_np(scores_gamma, names, keys))
```
